path to image

```python
IMAGE_PATH = '/content/seed0016.png' #path to image 
```

setup language and read image

```python
reader = easyocr.Reader (['ar'], gpu = True) 
result = reader.readtext(IMAGE_PATH)
result
```

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(IMAGE_PATH)
img = cv2.rectangle(img, top_left, bottom_right, (0, 255, 0), 5)
img = cv2.putText(img, text, top_left, font, .5, (255,255,255), cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

##multiple lines

```python
img = cv2.imread(IMAGE_PATH)
for detection in result:
  top_left = tuple([int(val) for val in detection[0][0]])
  bottom_left = tuple([int(val) for val in detection[0][2]])
  text = detection[1]
  font = cv2.FONT_HERSHEY_SIMPLEX
  img = cv2.rectangle(img, top_left, bottom_right, (0, 255, 0), 5)
  img = cv2.putText(img, text, top_left, font, 2, (255,255,255), cv2.LINE_AA)

plt.figure(figsize = (10,10))
plt.imshow(img)
plt.show()
```
